Seaborn HeatMap

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr_2d = np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(data=arr_2d)
```

```python
df = pd.read_csv("Who_is_responsible_for_global_warming.csv")
df
```

```python
df=df.drop(columns=["Country Code","Indicator Name","Indicator Code"], axis=1).set_index("Country Name")
df.head(6)
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(data=df,cmap = "coolwarm", annot = True)
```
