```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar_men = [112,48,89,85,94,89,124,78,105,76,120,90]
blood_sugar_women = [114,88,98,85,98,109,74,108,109,78,110,80]
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 160], rwidth=0.95, 
         label=['Men', 'Women'], color=['blue', 'orange'])
plt.title("Blood Sugar Levels")
plt.xlabel("Blood Sugar Range")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
