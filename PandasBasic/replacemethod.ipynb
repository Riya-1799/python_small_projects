```python
import pandas as pd
import numpy as np
df= pd.read_csv("weather_data_replacefun.csv",parse_dates=["day"])
df
```

```python
df['temperature'] = df['temperature'].replace(r'[A-Za-z]', '', regex=True)
df['windspeed'] = df['windspeed'].replace(r'[A-Za-z]', '', regex=True)
df
```

```python
df['temperature'] = pd.to_numeric(df['temperature'], errors='coerce')
df['windspeed'] = pd.to_numeric(df['windspeed'], errors='coerce')
```
