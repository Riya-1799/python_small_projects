```python
import pandas as pd
import numpy as np
df= pd.read_csv("weather_data_replacefun.csv")
df
```

```python
new_df = df.replace(-99999,np.nan)
new_df
```

```python
new_df=df.replace({
    'temperature' : -99999,
    'windspeed' : -99999,
    'event' : '0'
},np.nan)
new_df
```

```python
new_df = df.replace({
    -99999: np.nan,
    'No event' : 'Sunny'
})
new_df
```

```python
new_df = df.replace('[A-Za-z]','', regex=True)
new_df
```

```python
new_df=df.replace({
    'temperature' :'[A-Za-z]',
    'windspeed' : '[A-Za-z]'
},'',regex=True)
new_df
```
