```python
import pandas as pd
df1 = pd.DataFrame({
    "city" : ["mumbai","ahm","delhi"],
    "humadity" : [60,40,65]
})
df1  
```

```python
import pandas as pd
df2 = pd.DataFrame({
    "city" : ["ahm","delhi","Nadiad"],
    "temprature" : [40,35,38]
    
})
df2 
```

```python
df3=pd.merge(df1,df2,on="city",how="outer")
df3
```

```python
df4=pd.merge(df1,df2,on="city",how="inner")
df4
```

```python
df5=pd.merge(df1,df2,on="city",how="left")
df5
```

```python
df6=pd.merge(df1,df2,on="city",how="right")
df6
```
